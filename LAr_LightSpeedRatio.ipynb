```python
import numpy as np
import matplotlib.pyplot as plt

a0  = 1.26
aUV = 0.23
aIR = 0.0023
lUV = 106.6
lIR = 908.3
lSc = 128

def v_ratio(l):
    num = a0 + (aUV*lSc**2) / (lSc**2 - lUV**2) + (aIR*lSc**2) / (lSc**2 - lIR**2)
    dem = a0 + (aUV*l**2) / (l**2 - lUV**2) + (aIR*l**2) / (l**2 - lIR**2)
    return np.sqrt(num/dem)
    
    
    
```

```python
lrange = np.linspace(128,1000,100)
plt.plot(lrange,v_ratio(lrange))
plt.ylabel(r'$v_{\rm Ch} (\lambda_{\rm Ch})\; / \; v_{\rm Sc} (\lambda_{\rm Sc} = 128\;{\rm nm})$',fontsize=14)
plt.xlabel(r'Cherenkov wavelength $\lambda_{\rm Ch} \; [{\rm nm}]$',fontsize=14)
plt.savefig('LAr_LightSpeedRatio.png')
plt.show()
```
