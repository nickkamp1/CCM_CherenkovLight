```python
import numpy as np
import matplotlib.pyplot as plt
# import LAr stuff
from VisualizationTools.geant4_detector_specs import lar_Energy_rin,lar_Energy_rs,lar_Energy_scin
from VisualizationTools.geant4_detector_specs import lar_SCINT,lar_RIND,lar_RSL
from VisualizationTools.geant4_detector_specs import DefineLAr,hc
# import TPB stuff
from VisualizationTools.geant4_detector_specs import TPBEnergy,TPBEmission,TPBRIndex,DefineTpb

```

```python
plt.plot(hc/np.array(lar_Energy_scin),lar_SCINT)
plt.xlabel('Wavelength [nm]')
plt.ylabel('Scintillation Spectrum')
plt.savefig('Figures/G4_Detector/ScintSpectrumAr.png')
plt.show()

fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(hc/np.array(lar_Energy_rin),lar_RIND,'g-')
ax2.plot(hc/np.array(lar_Energy_rs),lar_RSL,'b-')
ax1.set_xlabel('Wavelength [nm]')
ax1.set_ylabel('Refractive Index', color='g')
ax2.set_ylabel('Rayleigh Scattering Length [cm]', color='b')
plt.semilogy()
plt.savefig('Figures/G4_Detector/Ridx_RayleighScat_Ar.png')
plt.show()

lar_Energy_abs,lar_ABSL,lar2_ABSL = DefineLAr()

fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(hc/np.array(lar_Energy_abs),lar_ABSL,'g-')
ax2.plot(hc/np.array(lar_Energy_abs),lar2_ABSL,'b-')
ax1.set_xlabel('Wavelength [nm]')
ax1.set_ylabel('Abs Length Nominal LAr [cm]', color='g')
ax2.set_ylabel('Abs Length Murky LAr [cm]', color='b')
ax1.semilogy()
ax2.semilogy()
ax1.set_ylim(1e0,5e3)
ax2.set_ylim(1e0,5e3)
plt.savefig('Figures/G4_Detector/AbsLength_Ar.png')
plt.show()
```

```python
plt.plot(hc/np.array(TPBEnergy),TPBEmission)
plt.xlabel('Wavelength [nm]')
plt.ylabel('Emission Spectrum')
plt.savefig('Figures/G4_Detector/EmitSpectrumTPB.png')
plt.show()

plt.plot(hc/np.array(TPBEnergy),TPBRIndex)
plt.xlabel('Wavelength [nm]')
plt.ylabel('Refractive Index')
plt.savefig('Figures/G4_Detector/Ridx_TPB.png')
plt.show()

TPBWLSAbsorption,TPBWLSAbsorption100,TPBAbsorption = DefineTpb()

plt.plot(hc/np.array(TPBEnergy),TPBWLSAbsorption,label='TPBWLSAbsorption (Foils)')
plt.plot(hc/np.array(TPBEnergy),TPBWLSAbsorption100,label='TPBWLSAbsorption100 (PMTs)')
plt.plot(hc/np.array(TPBEnergy),TPBAbsorption,label='TPBAbsorption')
plt.xlabel('Wavelength [nm]')
plt.ylabel('Absorption Length [cm]')
plt.semilogy()
plt.legend()
plt.xlim(0,1000)
plt.savefig('Figures/G4_Detector/AbsLength_TPB.png')
plt.show()
```
